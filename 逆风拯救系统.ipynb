# 神秘导引


啊啊啊，不甘心啊，就这么失败了！戴上电子手铐脚链的那一刻，维妮娜彻底的后悔了。为什么自己要故意去针对罗峰呢，自己之前为什么就那么眼高手低，没有识别出他位面之子的超级气运，又或者是因为注定的自己的作为主角垫脚石的设定，想到这里，坐在徐徐起飞前往旧世界的智能战机上，维妮娜终于流出了不甘的泪水，不过还没有带泪水挤出眼眶，他别用手揩了锴眼角，将那一点点伤心的痕迹抹去，或许按照他说的那样，自己这样做了，一切就不会发生，这一切还有改变，到了这个时候，一切应该还有转机…….

-----

时间倒回到昨天凌晨三点三十，江南在押高级战神级武者监狱A4031号监狱，沉溺了许久痛苦和疯狂的维妮娜还在思考中，突如其来，一切都静止了，风好像也停了，天好像也亮了，监狱隔壁舍友的鼾声好像也听不见了，但这一切维妮娜那好像都没有注意到，默默地，这一切都发生了。直到刚刚，空间中凭空出现的巨大亮光，才让唯一那停止了思考，虽然武器被收缴，但是作为战神级武者他战斗意识立刻让他反应过来，这一切不寻常，于是他拼尽全力站了起来，但他唯一能做到的也仅仅如此，这才让他意识到一切不同寻常！

“嗨，维妮娜美女，你好啊！”，凭空之中，空气中传来了一阵声音，“作为被我选中的第一个女反派，我决定第一个拯救的就是你”。


空气中突如其来的就传来了这段声音，虽然他身体拼尽全力无法动弹，但是刹那之间高级战神级的思考力却让他诧异万分，“什么鬼啊！我可是高级战神，你是谁敢对我不敬？等我出了这监狱看我不搞死你……”下意识地，处于战神级强者的骄傲与傲慢，立刻就愤怒的想到。

“呵呵，小妞，你自我感觉还不错呀，可惜你没有这个机会了，你出了这监狱下一步就是你的停尸场，估计你没有什么机会了”，“什么？……”，虽然早有预感，但是这个东西也只是战神宫内部的绝密资料，而且目前审判的结果也没下来，为什么这声音那么肯定，这一定不对劲！“嗨，小妞呀，你别急呀！ 点个同意，我让你亲自看到你自己的命运，以及为啥我会这样说，同意了之后你都会懂得”，光束中神秘的声音这样说，“别害怕，我不会害你的，要是害你，我不来，你明天也就凉了。”，听到那个声音那样说道，维妮娜想着自己反正也没有其他选择，动也动不了，那就点个同意吧，看他能怎么样！于是就想到怎么点同意呀！只不过在他这个念头刚出来的一瞬间，他的意识就离奇地传送到了一个非常神秘的空间！

恍惚之间，光影颤动。流光溢彩，玲珑剔透。忽然之间自己就来到这空间，只看到对面一个神秘的男子，而且自己也自由了，并且令他欣喜的就是，自己的能力也已经全部能用了，不仅仅是精神方面的被禁锢的能力，还有这一些进监狱之前被没收掉的装备和武器，都已经给自己装备好了，而且还是自己的最强状态，满弹满装药。只见此时的维妮娜，英姿飒爽，意气风发，一体式高科技纳米纤维作战服，右手长达数十公分的顶阶克罗合金战刀，左手SSA级高射速机枪腕带，超级AI智能辅助瞄准作战头盔，一只油光锃亮的作战靴，满级的装备以及感受到的空间腕带里的满满的弹药，顿时让她自信满满。

于是他下意识的抬起了机枪，对着面前的人影就扣下了扳机。令他意外的是，机枪如愿以偿的激发了，钢铁洪流一般的弹幕笼罩了那个人的背影，但更让他惊奇的就是，尽管有着轰鸣如雷的子弹洪流，但是那个人的声音还是准确的传的过来。“呵呵，你是不可能伤害我的。”， 那人轻蔑地说道。“愚蠢的女人啊，用心想想怎么可能在我的领域击败我自己。恢复装备只是给你的一个体验礼物而已，不要白费力气了，静静的听我说吧。”钢铁洪流肆虐而过，自动换电的看看生终于让这个空间暂停了几许，之间，那人只是简简单单的一挥手，好像一切又停止了！


“傻女人啊，给你看看你的命运吧。这是那吞噬星空原著小说，里面的一切都是你的命运，看看吧。”，不容分说之下，那本金黄色的书就冲着维妮娜飞了过来，不由分说就撞进了维妮娜的脑袋。唰唰唰唰伴随着熟悉的快速翻页的声音，空间再次陷入了安静！“怎么可能啊”，维妮娜震惊地说道，“我的命运是注定了的吗？后面的结局是这样，也就是说我马上会死，而且是曝尸荒野？我不相信。”


“呵呵，不相信就对了，我想但凡是个人都很难相信这一点。但是这一切对于你来说就是事实啊，而且这结局离你已经很近了，相信我吧，最晚明天晚上，如果你不根据我说的去做，你的结局就像书中的那样，而这一切对我毫无影响。”，那个神秘的声音说道。“出于特殊的原因，我特意消耗巨大的资源，浪费大量的时间来这个空间拯救你，不过这一切的选择权都在你，选择相信我或者不相信，这一切都在于你，你自己看着办吧。”说完了那句话，那个背影就开始转过身来，结果维妮娜惊讶的发现，在他的事业中，那个背影也只是一团黑雾而已。“这样吧，给你足够的选择时间，根据我的调查，你在这个世界的有效存活时间已经不足于24小时了，而跟随我的选择，你便可以跳脱这个世界，具体的选择权还是在你吧，接下来的时间里，你有这三次安全的机会来召唤我。想好了，到时候再联系我吧，这次你处的空间有时空暂停的功能，而此刻你拥有的暂停时间还有十分钟，有什么问题的话可以问我，我可以为你回答”，那个黑影说到，重新转身过去。

突如其来接收到大量的信息，维妮娜顿时感觉很是懵逼。即使是出于高级战神级的大脑，大量的信息他都来不及消化，让她陷入了懵逼的状态。“你是什么情况啊？到现在为止你还没介绍你自己呢，而且你对我好像很了解的样子”，下意识的，维妮娜想到！

“哎呀，好尴尬呀，说了这么久，结果忘记自己介绍了。可能是我之前太过内向了吧，很少做自我介绍的原因。或许这一切应该做个改善…”，那黑影嘟囔到，“但是这一切都不要紧，反正在这个世界里，我就是全能的。”，“哈喽，你好啊，刚才忘记给你介绍了，那么就重新介绍一下吧。刚看了一下时间可能就剩十几分钟了，那我就简短介绍吧。这个世界你的活动也自由，我就不限制了，但是你也不要浪费你的资源。毕竟这一切你都能拿回去自己用的呢，而且在这里你的伤害对我也是无效的。其实关于我的存在，我表示自己都说不清楚。 突然间就可以做到这一点，看着看着小说，突然间就发现自己眼前出现了选择拯救女反派的按钮，出于自己内心那种致命的同情心吧，我点了是就过来了，然后也就出现了在你这儿，也没有别的东西，其他的我也不清楚，毕竟这一切的一切选择权都不在我呀。”，听着那声音，那黑影的也是很委屈，仿佛就像一个就是旧世界一个很普通的弱者，维妮娜不屑的想到。

“嗯，没错，在你的认知里可能我就是个弱者，但是其实在我们那个世界里，我也只是个普通人而已。每天闲着无聊上上班儿，上班的时候总想着划水摸鱼，然后下班，作为单身狗的我也没有去找女朋友，看着小说，动漫事业发挥自己的想象，然后有朝一日，眼前出现了这个按钮，以为就是幻想，点了是，然后就到了这今天这一步，而你的出现也只是我的第一个尝试，或许你拒绝之后，我这个幻想就结束了吧，而按照书中的轨迹，你也只是一个简单的女反派，很快就会挂掉在长达千万字的主角的人生中，真的连炮灰都算不上。”

在这个空间中，那黑影好像具有强大的权限，基本上可以瞬间读取维妮娜的任何想法，基本上算是达到了无延迟单向对话。

“出现这个系统的时候，我好像在看一个动漫吧，然后一个关于吞噬星空的动漫，然后不经意间，就想到了主角晋升之前最大的女反派也就是你，然后也就简单的想了一下，点了个确定就来到这里，但是与你不一样的是，我可以随时离开这里，即使你死了，对我而言好像也不会有任何的损失。但是好像凭我的一己之力可以帮你改变一下你自己的人生，就这样，一切信息都是凭空出现的，我也是凭空接收而来的。我算是把所有的信息都给你了，下面的话，你有啥想问的，继续问吧，时间还有很多！”，那个黑影好像也在摸索着交流，但这时已经找到了方法。

“那么我想问的是，我的人生我差不多已经看完了，那么我要怎么改变呢？好像现在的我，这装备啥的都好像被没收了吖，而且根据书中的描写，明天审判结束，我就被押送要到荒野执行死刑，一切都是不可改变的呀。”维妮娜好像明白了一些，而且那个书吧，关于自己的，后面的章节也就一点点东西而已，简单的一扫而过，脸上的绝望的疑云变得更加深沉。

“其实我这个也是不知道的呀。刚才我才发现，我好像有能力改变你们这个世界的一些东西，但是这一切，都需要你的尽力配合才行。因为我来的之前接收到的第一个信息就是做的这一切，必须需要你的极力配合，具体到底怎么做吧，我考虑好了再跟你说吧。”

# 致命一刀


在执行专员看来，任务就是一瞬间。快刀斩过，完美的切断了罪犯的生机，任务完成应该去执行另一人的死刑了。于是头也不回，迅速回到了智能战机起飞离开，这一切用时甚至没有超过30秒，因为他们早已知道，在这个世界，即使斩断生机，如果一个人的生存欲望还是很强烈的话，这个人还是可以挣扎很久，但是一切的挣扎都是徒劳的，而且他们都有特定的任务，所以这一切都无需等待！

当快刀穿过身体的那一瞬间，时间仿佛一切都停止了。
